```python
# import data 
train = pd.read_csv('gdrive/My Drive/Humana/train_sampling.csv').copy()
train = train.iloc[: , 1:]
train.head()
```

```python
# Imputation using mean
colname = train.columns.tolist()
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
train_imputed = imp.fit_transform(train)
train = pd.DataFrame(train_imputed, index = None, columns = colname)
train.head()
```

```python
#reading data
X = datadrop(train,'covid_vaccination',1)
Y = train['covid_vaccination']

X.head()
```

```python
parameters_grid_ada = {
    'base_estimator':[DecisionTreeRegressor(max_depth = 3),DecisionTreeRegressor(max_depth = 4),DecisionTreeRegressor(max_depth = 5)],
    'learning_rate': [1e-5,1e-6,1e-7,1e-8,1e-9],
    'loss':['linear', 'square', 'exponential'],
    'n_estimators': np.arange(5,600,10)
}

ada_boosted_regressor = AdaBoostRegressor()
random_search_ada = RandomizedSearchCV(estimator = ada_boosted_regressor, 
                       param_distributions = parameters_grid_ada,
                       n_iter = 10, random_state = 5, cv = 5, n_jobs = -1, verbose = 2)
```
